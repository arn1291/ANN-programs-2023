```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bin_sig(x):
  return 1/(1+np.exp(-x))

def bipolar_sig(x):
  return (1-np.exp(-x))/(1+np.exp(-x))

def bipolar_step(x, theta):
  return np.where(x>theta, 1, -1)

def ramp(x):
  return np.where(x>0, x, 0)

x = np.linspace(-10, 10, 100)
print(x)
plt.plot(x, bin_sig(x))
plt.show()

plt.plot(x, bipolar_sig(x))
plt.show()

plt.plot(x, bipolar_step(x, 1))
plt.show()

plt.plot(x, ramp(x))
plt.show()
```
