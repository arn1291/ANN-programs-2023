```python
import numpy as np
import matplotlib.pyplot as plot

```

```python
x = np.linspace(-10, 10, 100)
print(x)
```

```python
def binarysigmoid(x):
  return 1 / (1 + np.exp(-x))
plot.plot(x, binarysigmoid(x))
plot.show()
```

```python
def bipolarsigmoid(x):
  return (1 - np.exp(-x)) / (1 + np.exp(-x))
plot.plot(x, bipolarsigmoid(x))
plot.show()
```

```python
def bipolarstep(x, theta):
  return np.where(x > theta, 1, -1)
plot.plot(x, bipolarstep(x, 1))
plot.show()
```

```python
def ramp(x):
  return np.where(x > 0, x, 0)
plot.plot(x, ramp(x))
plot.show()
```
