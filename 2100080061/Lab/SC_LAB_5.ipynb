```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def binary_sigmoid(x):
    return np.where(x > 0, 1, 0)

def bipolar_sigmoid(x):
    return np.where(x > 0, 1, -1)

y1 = binary_sigmoid(x)
y2 = bipolar_sigmoid(x)

plt.plot(x, y1, label='Binary Sigmoidal Function')
plt.plot(x, y2, label='Bipolar Sigmoidal Function')
plt.title("Sigmoidal Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```
